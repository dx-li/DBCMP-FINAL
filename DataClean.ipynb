```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv("WBData.csv")
df = df.dropna()
df = df.drop(df.iloc[:, -3:].columns, axis = 1).drop(["Country Code", "Series Code"], axis = 1)
```

```python
df = df.melt(id_vars = ["Country Name", "Series Name"], var_name = "Date", value_name = "Value")
df['Date'] = df['Date'].str[:4]
```

```python
df = df.pivot_table(values = "Value", index = ['Country Name', 'Date'], columns = ['Series Name'])
```

```python
df.to_csv("DataSet.csv")
```
